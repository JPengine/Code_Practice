```python
#Example 4.7
import numpy as np
import pandas as pd
import math

def simBM(n,T):
    times=np.linspace(0,T,num=n+1) # generate sequence
    Z=np.random.normal(loc=0.0, scale=1.0, size=n)  # generate a standard normal Z
    W=[0 for i in range(n)]  # list with 0 
    sigma=np.sqrt(np.diff(times)) #  compute sigma_k=sqrt(t_k - t_k-1)
    for k in range(1,n):
        W[k]=W[k-1]+sigma[k]*Z[k]
        
    return W
```

```python
# Generate three independent BM on the interval [0,1]
import random
random.seed(1)
n=200
x1=simBM(n,1)
x2=simBM(n,1)
x3=simBM(n,1)

import matplotlib.pyplot as plt
plt.plot(np.linspace(0,n,num=200),x1)
plt.plot(np.linspace(0,n,num=200),x2)
plt.plot(np.linspace(0,n,num=200),x3)
plt.show()
```
